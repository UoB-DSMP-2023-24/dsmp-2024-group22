```python
import pandas as pd
import numpy as np
from backtesting import Backtest, Strategy
from backtesting.lib import crossover
from backtesting.test import SMA, GOOG
```

```python
df=pd.read_excel(r"C:\Users\ROG\Desktop\so this is the sadness\MINI Project\Data\pred_df.xlsx",index_col=0)
df=df.dropna()
df=df.sort_values(by='Date')
df.rename(columns={'max_price': 'High', 'min_price': 'Low','open_price': 'Open','close_price': 'Close'}, inplace=True)
df
```

```python
df_predict=pd.read_excel(r"C:\Users\ROG\Desktop\so this is the sadness\MINI Project\Data\pred_df.xlsx")
df_predict['Open'] = pd.to_numeric(df_predict['Open'], downcast='float')
df_predict['High'] = pd.to_numeric(df_predict['High'], downcast='float')
df_predict['Low'] = pd.to_numeric(df_predict['Low'], downcast='float')
df_predict
```

```python
df_predict=df_predict.set_index("Date")
df_predict
```

```python
class SmaCross(Strategy):
    n1 = 1
    n2 = 5
    def init(self):
        close = self.data.Close
        self.sma1 = self.I(SMA, close, self.n1)
        self.sma2 = self.I(SMA, close, self.n2)

    def next(self):
        if crossover(self.sma1, self.sma2):
            self.buy()
        elif crossover(self.sma2, self.sma1):
            self.sell()
bt = Backtest(df_predict, SmaCross,
              cash=10000, commission=.002,
              exclusive_orders=True)

output = bt.run()
output
```
