```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import seaborn as sns

folder_path = r'C:\Users\tlext\Desktop\Group Project'

csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

max_volume = []

for file in csv_files:
    file_path = os.path.join(folder_path, file)  # Construct the full file path
    df = pd.read_csv(file_path)  # Read the CSV file into a DataFrame
    max_sold = df.iloc[:, 2].max()  # Find the max value in the third column
    max_volume.append(max_sold)

print(max_volume)

min_volume = []

for file in csv_files:
    file_path = os.path.join(folder_path, file)  # Construct the full file path
    df = pd.read_csv(file_path)  # Read the CSV file into a DataFrame
    min_sold = df.iloc[:, 2].min()  # Find the max value in the third column
    min_volume.append(min_sold)

print(min_volume)





quantity_sold = []


for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    quantity_sold.extend(df.iloc[:, 2])  


plt.figure(figsize=(10, 6))
plt.hist(quantity_sold, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Quantity Sold')
plt.xlabel('Quantity Sold')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()



```

```python
modes = []


for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    mode_sold = df.iloc[:, 2].mode().values[0] 
    modes.append(mode_sold)


plt.figure(figsize=(10, 6))
plt.bar(range(len(csv_files)), modes, color='skyblue')
plt.xticks(range(len(csv_files)), csv_files, rotation=45, ha='right')
plt.title('Mode Selling Quantity for Each File')
plt.xlabel('File')
plt.ylabel('Mode Selling Quantity')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
from tqdm import tqdm
dfs = []

for file in tqdm(csv_files):
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path, names=['Time', 'Price', 'Quantity'])
    dfs.append(df)

merged_dfs = pd.concat(dfs, ignore_index=True)

print(merged_dfs.head())
print(merged_dfs.info())

```

```python
plt.figure(figsize=(10, 6))
plt.hist(merged_dfs['Price'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
## SUMMARY STATISTICS ##
import numpy as np

mean_price = np.mean(merged_dfs['Price'])
print(f'MEAN PRICE OF THE STOCK: {mean_price}')

standard_dev_price = np.std(merged_dfs['Price'])
print(f'STANDARD DEV OF PRICE: {standard_dev_price}')

mean_quantity = np.mean(merged_dfs['Quantity'])
print(f'MEAN PRICE OF THE STOCK: {mean_quantity}')

standard_dev_quantity = np.std(merged_dfs['Quantity'])
print(f'STANDARD DEV OF PRICE: {standard_dev_quantity}')

```
