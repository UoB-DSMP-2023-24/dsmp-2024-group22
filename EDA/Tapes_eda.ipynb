```python
import numpy as np
folder_path = r'C:\Users\tlext\Desktop\Group Project'

csv_files = [file for file in os.listdir(folder_path) if file.endswith('.csv')]

max_prices = []

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    max_price = df.iloc[:, 1].max()
    max_prices.append(max_price)

print(max_prices)
print(len(max_prices))



max_price_df = pd.DataFrame({'Max Price': max_prices})

output_filepath = r'C:\Users\tlext\Desktop\Daily Max Price.csv'
max_price_df.to_csv(output_filepath, index=False)

min_prices = []

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    min_price = df.iloc[:, 1].min()
    min_prices.append(min_price)

print(min_prices)
print(len(min_prices))

min_price_df = pd.DataFrame({'Min Price': min_prices})

output_filepath = r'C:\Users\tlext\Desktop\Daily Min Price.csv'
min_price_df.to_csv(output_filepath, index=False)


max_min_df = pd.DataFrame({'Max Price': max_prices, 'Min Price': min_prices})
output_filepath = r'C:\Users\tlext\Desktop\Daily Max-Min Price.csv'
max_min_df.to_csv(output_filepath, index=False)




average_prices = []

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    average_price = np.mean(df.iloc[:, 1])
    average_prices.append(average_price)

print(average_prices)
print(len(average_prices))

average_price_df = pd.DataFrame({'Average Price': average_prices})

output_filepath = r'C:\Users\tlext\Desktop\Daily Average Price.csv'
average_price_df.to_csv(output_filepath, index=False)

max_min_av_df = pd.DataFrame({'Max Price': max_prices, 'Min Price': min_prices, 'Average Price': average_prices})
output_filepath = r'C:\Users\tlext\Desktop\Daily Max-Min-Av Price.csv'
max_min_av_df.to_csv(output_filepath, index=False)

spread_of_prices = []

for file in csv_files:
    file_path = os.path.join(folder_path, file)
    df = pd.read_csv(file_path)
    spread_of_price = df.iloc[:, 1].max() - df.iloc[:, 1].min()
    spread_of_prices.append(spread_of_price)

print(spread_of_prices)
print(len(spread_of_prices))

spread_df = pd.DataFrame({'Spread of Prices': spread_of_prices})
output_filepath = r'C:\Users\tlext\Desktop\\Spread of Prices.csv'
spread_df.to_csv(output_filepath, index=False)

price_spread = pd.read_csv(r'C:\Users\tlext\Desktop\Spread of Prices.csv')
plt.bar(range(len(spread_of_prices)), height=price_spread['Spread of Prices'])
plt.xlabel('Time')
plt.ylabel('Spread of Prices')
plt.title('Comparison Of The Spread Of Sell Prices Across All Tapes')
plt.show()


highest_prices = pd.read_csv(r'C:\Users\tlext\Desktop\Daily Max Price.csv')
sns.scatterplot(data=highest_prices)
plt.xlabel('Time')
plt.ylabel('Max Price Sold')
plt.show()

lowest_prices = pd.read_csv(r'C:\Users\tlext\Desktop\Daily Min Price.csv')
sns.scatterplot(data=lowest_prices)
plt.xlabel('Time')
plt.ylabel('Min Price Sold')
plt.show()

av_prices = pd.read_csv(r'C:\Users\tlext\Desktop\Daily Average Price.csv')
sns.scatterplot(data=av_prices)
plt.xlabel('Time')
plt.ylabel('Average Price Sold')
plt.show()

max_min_prices = pd.read_csv(r'C:\Users\tlext\Desktop\Daily Max-Min Price.csv')
sns.scatterplot(data=max_min_prices)
plt.xlabel('Time')
plt.ylabel('Max/Min Price Sold')
plt.yticks(np.arange(80, 420, 20))
plt.title('Max and Min Price Sold')
plt.show()

max_min_av_prices = pd.read_csv(r'C:\Users\tlext\Desktop\Daily Max-Min-Av Price.csv')
sns.scatterplot(data=max_min_av_prices)
plt.xlabel('Time')
plt.ylabel('Max/Min/Average Price Sold')
plt.yticks(np.arange(80, 420, 20))
plt.title('Max, Min and Average Price Sold')
plt.show()
```
